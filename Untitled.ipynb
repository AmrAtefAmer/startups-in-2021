```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#reading csv file
df = pd.read_csv('datasets/Startups in 2021 end.csv')
```

```python
#print first 5 rows to check data
df.head()
```

```python
#drop column that is not useful
df.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
#check shape of data 
df.shape
```

```python
#check columns and data types
df.info()
```

```python
#check colmns's data types
df.dtypes
```

```python
#chack sum of null values in each column
df.isnull().sum()
```

```python
#Convert data type for date from object to datetime
df['Date Joined'] = pd.to_datetime(df['Date Joined'],format=r'%m/%d/%Y',errors='ignore')
```

```python
#check again to see data type of Date Joined changed or not
df.dtypes
```

```python
#drop rows that contain null values 
df.dropna(inplace=True)
```

```python
#check null values again
df.isnull().sum()
```

```python
#create column contains company's start year
df['start_year'] = df['Date Joined'].dt.year
```

```python
#create column contains number of years for each compnay from start to now
df['num_years'] = 2021 - df['start_year']
```

```python
#removing $ from values of Valuation ($B) column to convert it to int
df['Valuation ($B)'] = df['Valuation ($B)'].apply(lambda x : x.replace('$',''))
```

```python
#convert Valuation ($B) column dtype to int
df['Valuation ($B)'] = df['Valuation ($B)'].astype(float)
```

```python
#change dtype of year to category 
df['start_year'] = df['start_year'].astype('category')
```

```python
#check dtypes again 
df.dtypes
```

```python
#convert dtype of industry column from object to category
df['Industry'] = df['Industry'].astype('category')
```

```python
df['Industry'].value_counts() #notice: here we see problem that 'Artificial Intelligence' appeared twice 
```

```python
#solution
df['Industry']=df['Industry'].apply(lambda x : x.capitalize())
```

```python
#"Artificial intelligence" problem solved but we see another problem "Fintech" and "Finttech"
df['Industry'].value_counts()
```

```python
#solution
df['Industry']=df['Industry'].apply(lambda x : x.replace('tt','t'))
df.groupby('Industry').sum()['Valuation ($B)']
```

# which company that have highest Valuation ?

```python
df[df['Valuation ($B)'] == df['Valuation ($B)'].max()]
```

#  in each country which company have highest Valuation?

```python
df.groupby('Country').max()['Company']
```

# Top 10 companies in Valuation

```python
df.sort_values('Valuation ($B)',ascending=False).head(10)['Country']
```

# average comapanies's valuation in each country ?

```python
df.groupby('Country').mean()['Valuation ($B)']
```

```python
df.groupby('Country').mean()['Valuation ($B)'].plot(kind='bar',figsize=(15,4))
plt.title("Average comapanies's valuation in each countr")
plt.show()
```

# Total comapanies's valuation in each country ?

```python
df.groupby('Country').sum()['Valuation ($B)']
```

```python
df.groupby('Country').sum()['Valuation ($B)'].plot(kind='bar',figsize=(15,4))
plt.title("total comapanies's valuation in each countr")
plt.show()
```

# number of companies in each year 
## here we can see the huge number of companies started in 2021

```python
df['start_year'].value_counts()
```

```python
df['start_year'].value_counts().sort_index(ascending = False).plot(kind='bar',figsize=(9,5))
plt.xlabel('Year')
plt.ylabel('Number of companies')
plt.title('Number of companies started in each year')
```

```python
df.start_year.dtype
```

# relation between total of Valuation for companies and start year

```python
df.groupby('start_year').sum()['Valuation ($B)'].plot(color='r')
plt.title('Total valuation for every year')
plt.ylabel('tota valuation')
plt.style.use('ggplot')
plt.show()
```

# relation between average of Valuation for companies and start year

```python
df.groupby('start_year').mean()['Valuation ($B)'].plot(color='g')
plt.title('average valuation for every year')
plt.ylabel('valuation average')
plt.show()
```

# Most of the industries that dominate companies

```python
df['Industry'].value_counts()
```

```python
#display with seaborn
sns.countplot(data=df,x='Industry')
plt.xticks(rotation=90)
plt.show()
```

```python
#display same data with matplotlib
df['Industry'].value_counts().plot(kind='bar',figsize=(10,4),color='c')
plt.xlabel('Industry')
plt.ylabel('number')
plt.show()
```

# industries that have most Valuations

```python
df.groupby('Industry').sum()['Valuation ($B)'].sort_values(ascending=False).plot(kind='bar',figsize=(15,4))
plt.xticks(rotation=90)
plt.show()
```

```python
# compare between number of years and total valuation 
```

```python
df.groupby('num_years').sum()['Valuation ($B)'].sort_values(ascending=False).plot(kind='bar',figsize=(20,4))
plt.xticks(rotation=90)
plt.show()
```

# number of industries in each country

```python
df.groupby('Country').count()['Industry'].plot(kind='bar',figsize=(15,4))
plt.title('number of industries in each country')
plt.show()
```

# number of each industry in each country

```python
h_map = pd.pivot_table(index='Industry',columns='Country',values='Valuation ($B)',data=df,aggfunc='count')
```

```python
plt.figure(figsize=(15,5))
sns.heatmap(h_map,annot=True)
plt.show()
```
